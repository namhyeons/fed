```python
cifar_10_num_params = [31.4, 178, 750, 1940] # K
cifar_100_num_params = [0.480, 1.35, 3.84, 12.3] # M

cifar_10_num_macs = [3.46, 23.6, 43.7, 62.6] # M
cifar_100_num_macs = [167, 310, 450, 585] # M

def select_level(id, levels, seed):
    
    i = id / 100
    
    if i < sum(levels[:1]):
        return 0
    elif i < (sum(levels[:2])):
        return 1
    elif i < (sum(levels[:3])):
        return 2
    else:
        return 3
```

```python
# interpolation - iid - # parameter

import matplotlib.pyplot as plt
import pickle
import numpy as np

with open("cifar10/iid[1]_K[1].pkl", 'rb') as f:
    interpolation_dict = pickle.load(f)

levels = interpolation_dict['levels']
x = []
y = interpolation_dict['accuracy']

for k in range(len(levels)):
    avg_params = 0.
    for i in range(100):
        level = select_level(i, levels[k], seed=0)
        avg_params += cifar_10_num_params[level]
    
    #print("{} / {} / {} / {} : {}, {}".format(levels[k][0], levels[k][1], levels[k][2], levels[k][3], avg_params / 10000, y[k]*100))
    avg_params /= 100
    x.append(avg_params)

index_min = min(range(len(x)), key=x.__getitem__)
index_max = max(range(len(x)), key=x.__getitem__)

print(len(x))

for i in range(len(x)):
    x[i] /= 100

z = np.polyfit(x, y, 1)
p = np.poly1d(z)
print(np.corrcoef(x, y))

plt.figure(dpi=120)
plt.title("CIFAR10 IID")
plt.scatter(x, y, color='dimgray')
plt.plot([x[index_min],x[index_max]],[p(x[index_min]),p(x[index_max])], linestyle='--', label='correlation coefficient : 0.76')

plt.xlabel("Average number of Parameters (1e5)") 
plt.ylabel("Test Accuracy")
plt.legend(loc='lower right')
plt.show()
```
